```python
# Scrap data from the official website (Presentation)
import requests 
from bs4 import BeautifulSoup

url = "https://www.esi-sba.dz/fr/index.php/presentation/"

response = requests.get(url)

soup = BeautifulSoup(response.content, "lxml")

docs = []
presentation = ""

paragraphs = soup.find_all("p")[:-3]
for p in paragraphs:
    presentation+='\n' + p.get_text()
    

span_elements = soup.find_all("span", class_="gdlr-core-icon-list-content")[:4]
spans = ""
for s in span_elements:
    spans+='\n' + s.get_text()

presentation+=spans
print(presentation)

docs.append({
    "url": url,
    "content": presentation
})
```

### Load JSON as Langachain Documents

```python
from langchain_community.document_loaders import JSONLoader
from langchain_core.documents import Document
from typing import List
import logging

loader = JSONLoader(
        file_path="../data/raw/ESI-SBA.json",
        jq_schema=".[]",
        content_key="content",
        metadata_func=lambda record, metadata: {
            "url": record.get("url"),
            **metadata,
        }
    )
documents = loader.load()
documents[0]
```

```python
max_tokens = 0
texts = [doc.page_content for doc in documents]

for text in texts:
    length = len(text.split())
    if length > max_tokens:
        max_tokens = length

print(max_tokens)
```
